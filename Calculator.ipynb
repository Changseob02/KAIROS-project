```python
import sympy as sp
from sympy.parsing.sympy_parser import standard_transformations, implicit_multiplication_application

def get_user_input():
    expression_str = input("식을 입력하세요: ")
    return expression_str

def main():
    x = sp.Symbol('x')
    transformations = (standard_transformations + (implicit_multiplication_application,))
    # transformations는 변환할 규칙들을 포함

    print("계산기 메뉴:")
    print("1. 사칙연산")
    print("2. 정적분")
    print("3. 부정적분")
    print("4. 행렬 연산")
    print("5. 미분")
    choice = input("원하는 작업을 선택하세요 (1/2/3/4/5): ")

    if choice == '1':
        expression_str = get_user_input()
        expression = sp.parse_expr(expression_str, transformations=transformations)

        try:
            result = expression.evalf()
            print(f"결과: {result}")
        except Exception as e:
            print("오류가 발생했습니다:", e)
    elif choice == '2':
        expression_str, lower_limit, upper_limit = get_user_input(), float(input("하한 구간을 입력하세요: ")), float(input("상한 구간을 입력하세요: "))

        try:
            expression = sp.parse_expr(expression_str, transformations=transformations)
            integral = sp.integrate(expression, (x, lower_limit, upper_limit))
            print(f"결과: {integral}")
        except Exception as e:
            print("오류가 발생했습니다:", e)

    elif choice == '3':
        expression_str = get_user_input()

        try:
            expression = sp.parse_expr(expression_str, transformations=transformations)
            antiderivative = sp.integrate(expression, x)
            print(f"결과: {antiderivative} + C")  # C는 적분 상수입니다.
        except Exception as e:
            print("오류가 발생했습니다:", e)

    elif choice == '4':
        matrix_str = input("행렬을 입력하세요 (예: [[1, 2], [3, 4]]): ")

        try:
            matrix = sp.Matrix(eval(matrix_str))  # 입력된 문자열을 파이썬 리스트로 변환
            print("행렬 연산을 선택하세요:")
            print("a. 덧셈")
            print("b. 뺄셈")
            print("c. 곱셈")
            matrix_choice = input("원하는 행렬 연산을 선택하세요 (a/b/c): ")

            if matrix_choice == 'a':
                other_matrix_str = input("다른 행렬을 입력하세요: ")
                other_matrix = sp.Matrix(eval(other_matrix_str))
                result = matrix + other_matrix
            elif matrix_choice == 'b':
                other_matrix_str = input("다른 행렬을 입력하세요: ")
                other_matrix = sp.Matrix(eval(other_matrix_str))
                result = matrix - other_matrix
            elif matrix_choice == 'c':
                other_matrix_str = input("다른 행렬을 입력하세요: ")
                other_matrix = sp.Matrix(eval(other_matrix_str))
                result = matrix * other_matrix
            else:
                print("유효하지 않은 입력입니다.")
                return

            print("결과:")
            print(result)

        except Exception as e:
            print("오류가 발생했습니다:", e)
    elif choice == '5':
        expression_str = get_user_input()

        try:
            expression = sp.parse_expr(expression_str, transformations=transformations)
            derivative = sp.diff(expression, x)
            print(f"미분 결과: {derivative}")
        except Exception as e:
            print("오류가 발생했습니다:", e)

if __name__ == "__main__":
    main()
```
